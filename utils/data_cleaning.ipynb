```python
with open('cleaned.txt', 'r') as file:
    lines = file.readlines()
```

```python
dhivehi_parts = []
english_parts = []

for line in lines:
    eng_part = line.split(',')[-1].split('|')[0].strip()
    dv_part = line.split('|')[-1].strip()

    dhivehi_parts.append(dv_part)
    english_parts.append(eng_part)
```

```python
import pandas as pd

df = pd.DataFrame({
    'latin': english_parts,
    'dhivehi': dhivehi_parts,
})
df.to_csv('training_data.csv', index=False)
```
